```python
# prompt: Get oribital velocity of galactic buldge as functions of radius

import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 4.300e-6  # Gravitational constant in kpc km^2/s^2 Msun^-1
M_bulge = 1e10  # Mass of the galactic bulge in Msun

# Radius range
r = np.linspace(0.1, 5, 100)  # Radius in kpc

# Calculate orbital velocity
v_circ = np.sqrt(G * M_bulge / r)

# Plot the result
plt.plot(r, v_circ)
plt.xlabel('Radius (kpc)')
plt.ylabel('Orbital Velocity (km/s)')
plt.title('Orbital Velocity of Galactic Bulge')
plt.grid(True)
plt.show()
```
